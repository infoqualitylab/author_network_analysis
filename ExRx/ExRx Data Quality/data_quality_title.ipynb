```python
#merge the dataframes to validate the names retrieved from scopus 
merged_df = pd.merge(left_df, right_df, left_on="Our ID", right_on="ID", how="left")
```

```python
merged_df.head()
```

```python
# Check the title names
for i in range(len(merged_df)):
    if merged_df.loc[i, 'Title_x'] != merged_df.loc[i, 'Title_y']:
        print("The title for article {} does not match".format(merged_df.loc[i, 'ID']))
        print("The title in the article article list is {}".format(merged_df.loc[i, 'Title_x']))
        
        print("The title in the autho info is {}".format(merged_df.loc[i, 'Title_y']))
        
    else:
        print("The title for article {} does match".format(merged_df.loc[i, 'ID']))
        
    
```
